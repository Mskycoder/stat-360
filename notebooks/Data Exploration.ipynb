```python
dfs = {}

for page_name in ['lead']:
    if SEND_REQUESTS:
        all_data = []
        
        for page_num in range(12000, 14001, 2):
            print(f"Requesting {page_name}, page {page_num}")
            r = requests.get(f'https://hackathon-api.bdc.n360.io/{page_name}?page={page_num}')
            data = json.loads(r.text)
            all_data.extend(data['Items'])

        dfs[page_name] = pd.DataFrame(all_data)
        print(f"Final Shape: {dfs[page_name].shape}")
        
        if SAVE:
            dfs[page_name].to_csv(f'../data/{page_name}.csv', index=False)
    
    else:
        dfs[page_name] = pd.read_csv(f'../data/{page_name}.csv')
```

```python
df = dfs['website_statistics']
segment_names = df.segment_name.value_counts().index.tolist()
website_id = 328

stats = ['avg_time_on_page']
stat = stats[0]

y_name = stat.replace('_', ' ').title()

traces = []

for segment_name in segment_names:
    start_date = '2017-01-01'
    end_date = '2020-01-01'
    
    segment_mask = df['segment_name'] == segment_name
    website_mask = df['website_id'] == website_id
    start_date_mask = df['date_start'] > start_date
    end_date_mask = df['date_start'] < end_date
    masked_df = df[segment_mask & website_mask & start_date_mask & end_date_mask]
    
    trace = go.Scatter(
        x=masked_df.date_start.values,
        y=masked_df[stat].values,
        name=segment_name
    )
    
    traces.append(trace)

layout = go.Layout(
    title=f"Monthly statistics for Website #{website_id}",
    xaxis=dict(title="Start Date"),
    yaxis=dict(title=y_name)
)
fig = go.Figure(traces, layout)

py.iplot(fig)
```

```python
col_name = 'lead_type'
lead_df = dfs['lead']
source_mask = lead_df['lead_source'] == 'Web'

masked_df = lead_df[source_mask]

pie_data = masked_df[col_name].value_counts()[:15]
```
